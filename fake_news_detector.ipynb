```python
import re
import nltk
from nltk.corpus import stopwords
from sklearn.model_selection import train_test_split

nltk.download('stopwords' ,  quiet=True)
nltk.download('punkt' ,  quiet=True)
nltk.download('punkt_tab' ,  quiet=True)

def clean_text(text):
  if isinstance(text,str) :
    text = text.lower()
    text = re.sub(r'[^a-zA-Z0-9\s]', '', text)
    text = re.sub('\s+', ' ',text)

    return text
  return ""

combined_df['clean_title'] = combined_df['title'].apply(clean_text)
combined_df['clean_text'] = combined_df['text'].apply(clean_text)
combined_df['clean_subject'] = combined_df['subject'].apply(clean_text)

stop_words = set(stopwords.words('english'))

def remove_stopwords(text) :
  words = nltk.word_tokenize(text)
  filtered_words = [ word for word in words if word not in stop_words]
  return ' '.join(filtered_words)

combined_df['clean_title'] = combined_df['clean_title'].apply(remove_stopwords)
combined_df['clean_text'] = combined_df['clean_text'].apply(remove_stopwords)
combined_df['clean_subject'] = combined_df['subject'].apply(remove_stopwords)

features = ['clean_title' , 'clean_text' , 'clean_subject']
target = ['status']

x = combined_df[features]
y = combined_df[target]

x['combined_text'] = x['clean_title'] + ' ' + x['clean_text'] + ' ' + x['clean_subject']

x_train , x_test , y_train , y_test = train_test_split(x , y , test_size = 0.2 , random_state = 42 )

print(f"training set size : {len(x_train)}")
print(f"testing set size : {len(x_test)}")
```

```python
print("\nTest set distribution:")
print(y_test['status'].value_counts())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer(
    max_features = 15000 ,
    max_df = 0.85 ,
    min_df = 2 ,
    ngram_range = (1,3) ,
    sublinear_tf = True
)
x_train['combined_text'] = x_train['clean_title'] + ' ' + x_train['clean_text'] + ' ' + x_train['clean_subject']
x_test['combined_text'] = x_test['clean_title'] + ' ' + x_test['clean_text'] + ' ' + x_test['clean_subject']

x_train_tfidf = tfidf_vectorizer.fit_transform(x_train['combined_text'])
x_test_tfidf = tfidf_vectorizer.transform(x_test['combined_text'])
```

```python
print(f"feature matrix shape : {x_train_tfidf.shape}")
print(f"no. of features : {len(tfidf_vectorizer.get_feature_names_out())}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC
from sklearn.naive_bayes import MultinomialNB

from sklearn.metrics import accuracy_score , classification_report , confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Logistic regression
lr_model = LogisticRegression(max_iter = 1000 , random_state = 42 )
lr_model.fit(x_train_tfidf, y_train)
lr_pred = lr_model.predict(x_test_tfidf)

lr_accuracy = accuracy_score(y_test, lr_pred)
print("Logistic Regression accuracy score : " , lr_accuracy)
print("Logistic Regression classification report : " , classification_report(y_test , lr_pred))
cm = confusion_matrix(y_test,lr_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Real', 'Fake'], yticklabels=['Real', 'Fake'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Logistic Regression')
plt.show()
```

```python
# random forest classifier
rf_model = RandomForestClassifier(random_state = 42)
rf_model.fit(x_train_tfidf , y_train)
rf_pred = rf_model.predict(x_test_tfidf )

rf_accuracy = accuracy_score(y_test, rf_pred)
print("Random Forest accuracy score : " , rf_accuracy)
print("Random Forest classification report : " , classification_report(y_test , rf_pred))
cm = confusion_matrix(y_test, rf_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Real', 'Fake'], yticklabels=['Real', 'Fake'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest')
plt.show()
```

```python
svm_model = LinearSVC(C=1.0, random_state=42, max_iter=1000)
svm_model.fit(x_train_tfidf, y_train)
svm_pred = svm_model.predict(x_test_tfidf)

svm_accuracy = accuracy_score(y_test, svm_pred)
print("SVM accuracy score:", svm_accuracy)
print("SVM classification report:", classification_report(y_test, svm_pred))
cm = confusion_matrix(y_test, svm_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Real', 'Fake'], yticklabels=['Real', 'Fake'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - SVM')
plt.show()
```

```python
nb_model = MultinomialNB()
nb_model.fit(x_train_tfidf , y_train)
nb_pred = nb_model.predict(x_test_tfidf)

nb_accuracy = accuracy_score(y_test, nb_pred)
print("naive bayes accuracy score : " , nb_accuracy)
print("naive bayes classification report : " , classification_report(y_test , nb_pred))
cm = confusion_matrix(y_test, nb_pred)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Real', 'Fake'], yticklabels=['Real', 'Fake'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - naive bayes')
plt.show()
```

```python
import pickle
import re
import string
import nltk
nltk.download('stopwords', quiet=True)
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

def preprocess_text(text):
    """Preprocess text for prediction"""
    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = re.sub(r'[^\w\s]', ' ', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove whitespaces
    text = text.strip()

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]

    return ' '.join(filtered_words)

def predict_news(news_text, model, vectorizer):
    """Predict if news is real or fake"""
    # Preprocess the text
    preprocessed_text = preprocess_text(news_text)

    # Transform using vectorizer
    text_vectorized = vectorizer.transform([preprocessed_text])

    # Make prediction
    prediction = model.predict(text_vectorized)[0]

    # Return result based on numeric encoding (1=fake, 0=real)
    if prediction == 1:
        return "FAKE NEWS"
    else:
        return "REAL NEWS"
```

```python
%%writefile app.py

import streamlit as st
import pickle
import numpy as np
import re
import string
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk

# Download necessary NLTK data
try:
    nltk.data.find('corpora/stopwords')
except LookupError:
    nltk.download('stopwords')

# Set page configuration
st.set_page_config(
    page_title="Fake News Detector",
    page_icon="🔍",
    layout="centered"
)

# Function to load saved models
@st.cache_resource
def load_models():
    # Load the trained model
    with open('fake_news_model.pkl', 'rb') as f:
        model = pickle.load(f)

    # Load the vectorizer
    with open('tfidf_vectorizer.pkl', 'rb') as f:
        vectorizer = pickle.load(f)

    return model, vectorizer

# Text preprocessing function
def preprocess_text(text):
    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    text = re.sub(f'[{string.punctuation}]', ' ', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Remove whitespaces
    text = text.strip()

    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    words = text.split()
    filtered_words = [word for word in words if word not in stop_words]

    # Apply stemming
    ps = PorterStemmer()
    stemmed_words = [ps.stem(word) for word in filtered_words]

    return ' '.join(stemmed_words)

# Main function to run the app
def main():
    # Load the model and vectorizer
    try:
        model, vectorizer = load_models()
        models_loaded = True
    except FileNotFoundError:
        st.error("Model files not found. Please ensure 'fake_news_model.pkl' and 'tfidf_vectorizer.pkl' are in the current directory.")
        models_loaded = False

    # App title and description
    st.title("📰 Fake News Detector")
    st.markdown("""
    This application uses machine learning to identify potentially fake news articles.
    Enter the text of a news article to see if it might be fake or reliable.
    """)

    # Create a text area for user input
    news_text = st.text_area("Enter News Text", height=200,
                            placeholder="Paste the news article text here...")

    # Create a button to trigger prediction
    if st.button("Analyze Text"):
        if not news_text:
            st.warning("Please enter some text to analyze.")
        elif models_loaded:
            with st.spinner("Analyzing..."):
                # Preprocess the text
                preprocessed_text = preprocess_text(news_text)

                # Transform text using the loaded vectorizer
                text_vectorized = vectorizer.transform([preprocessed_text])

                # Make prediction
                prediction = model.predict(text_vectorized)[0]

                # Display results
                st.header("Analysis Results")

                if prediction == 1:
                    st.error("⚠️ This text is classified as POTENTIALLY FAKE NEWS")

                    # Display key indicators of fake news
                    st.subheader("Key Indicators")
                    st.markdown("""
                    The analysis detected patterns commonly associated with misleading content:
                    - Sensationalist language and excessive claims
                    - Limited source attribution
                    - Inconsistent narrative structure
                    """)
                else:

                    st.success(f"✅ This text is classified as LIKELY RELIABLE")

                    # Display key indicators of reliable news
                    st.subheader("Key Indicators")
                    st.markdown("""
                    The analysis detected patterns commonly associated with reliable content:
                    - Balanced presentation of facts
                    - Proper source attribution
                    - Coherent narrative structure
                    """)

                # Add disclaimer
                st.info("Note: This is an automated analysis and should not be the sole factor in determining the reliability of news content. Always verify information from multiple credible sources.")

    # Add information about the model
    with st.expander("About this Model"):
        st.write("""
        This fake news detector uses a machine learning model trained on a dataset of labeled news articles.
        The model analyzes patterns in the text to identify features commonly associated with fake news.

        **Limitations:**
        - The model can only analyze text content, not images or videos
        - Performance depends on the training data used
        - Cultural and contextual nuances may not be fully captured
        """)

# Run the app
if __name__ == "__main__":
    main()
```
